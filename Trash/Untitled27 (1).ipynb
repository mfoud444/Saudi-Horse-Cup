```python
import csv

def read_csv(file_path):
    with open(file_path, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        data = list(reader)
    return data

def write_csv(file_path, data):
    with open(file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        for row in data:
            writer.writerow(row)

def merge_csv_files(file_paths, output_file_path):
    if not file_paths:
        raise ValueError("No CSV files provided to merge.")

    # Read the first file to get the headers and initial data
    merged_data = read_csv(file_paths[0])
    headers = merged_data[0]

    # Iterate through the remaining files and merge their contents
    for file_path in file_paths[1:]:
        data = read_csv(file_path)
        if data[0] != headers:
            raise ValueError(f"CSV file {file_path} has different headers.")
        merged_data.extend(data[1:])  # Append the data without headers

    # Write the merged data to the output file
    write_csv(output_file_path, merged_data)

# List of paths to the CSV files
csv_files = ['tweets1.csv', 'tweets2.csv', 'tweets3.csv','tweets4.csv', 'tweets5.csv','output.csv','1.csv']  # Add more file paths as needed
output_file_path = 'merged_tweets.csv'

# Merge the CSV files
merge_csv_files(csv_files, output_file_path)

print(f"Merged data has been written to {output_file_path}")
```
